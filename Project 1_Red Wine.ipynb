```python
a = df.drop('quality', axis = 1)
print(a)
```

```python
w = df['quality'].apply(lambda y_value:1 if y_value >=7 else 0)
w
```

```python
a_train, a_test, w_train, w_test = train_test_split(a,w, test_size = 0.3, random_state = 3)
```

```python
print(w.shape, w_train.shape, w_test.shape)
```

```python
print(a.shape, a_train.shape, a_test.shape)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(a_train, w_train)
```

```python
a_test_prediction = model.predict(a_test)
test_data_accuracy = accuracy_score(a_test_prediction, w_test)
```

```python
print('Accuracy : ', test_data_accuracy)
```
